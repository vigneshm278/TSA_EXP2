```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Dataset
data = pd.read_excel('/content/supermarket_10years.xlsx')

# Group yearly sales
resampled_data = data.groupby('Year')['Sales_Amount'].sum().reset_index()

years = resampled_data['Year'].tolist()
sales = resampled_data['Sales_Amount'].tolist()

# -----------------------------
# LINEAR TREND ESTIMATION
# -----------------------------

X = [i - years[len(years)//2] for i in years]
x2 = [i**2 for i in X]
xy = [i*j for i, j in zip(X, sales)]

n = len(years)

b = (n * sum(xy) - sum(sales) * sum(X)) / (n * sum(x2) - (sum(X)**2))
a = (sum(sales) - b * sum(X)) / n

linear_trend = [a + b * X[i] for i in range(n)]

print(f"Linear Trend Equation: y = {a:.2f} + {b:.2f}x")

# Plot Linear Trend
plt.figure(figsize=(8,5))

plt.plot(years, sales, marker='o', label='Actual Sales')
plt.plot(years, linear_trend, linestyle='--',
         label=f'Linear Trend: y={a:.2f}+{b:.2f}x')

plt.title('Linear Trend Estimation (Least Square Method)')
plt.xlabel('Year')
plt.ylabel('Sales Amount')
plt.legend()
plt.grid(True)
plt.show()


# -----------------------------
# POLYNOMIAL TREND ESTIMATION
# -----------------------------

coefficients = np.polyfit(X, sales, 2)

a0 = coefficients[2]
a1 = coefficients[1]
a2 = coefficients[0]

poly_trend = [a0 + a1*x + a2*(x**2) for x in X]

print(f"Polynomial Trend Equation: y = {a0:.2f} + {a1:.2f}x + {a2:.2f}x²")

# Plot Polynomial Trend
plt.figure(figsize=(8,5))

plt.plot(years, sales, marker='o', label='Actual Sales')
plt.plot(years, poly_trend, marker='o',
         label=f'Poly Trend: y={a0:.2f}+{a1:.2f}x+{a2:.2f}x²')

plt.title('Polynomial Trend Estimation - Degree 2 (Least Square Method)')
plt.xlabel('Year')
plt.ylabel('Sales Amount')
plt.legend()
plt.grid(True)
plt.show()
```
